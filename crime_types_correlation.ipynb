```python
df_street = pd.read_csv('correlation_data/2023-08-2025-03-street.csv', low_memory=False)
df_outcome = pd.read_csv('correlation_data/2023-08-2025-03-outcome.csv', low_memory=False)
```

```python
df_street = df_street.drop_duplicates()
df_outcome = df_outcome.drop_duplicates()
```

```python
substrings = ['theft', 'order', 'burglary', 'shoplifting', 'vehicle', 'damage', 'drugs', 'robbery']
# Make everything lowercase in Crime type in df_street
df_street['Crime type'] = df_street['Crime type'].str.lower()
# Make a new column in df_street called 'Substrings' that contains the substrings that are in the Crime type
def find_substrings(crime_type):
    found_substrings = [substring for substring in substrings if substring in crime_type]
    return ', '.join(found_substrings) if found_substrings else 'None'
df_street['Substrings'] = df_street['Crime type'].apply(find_substrings)
```

```python
if 'Month' in df_street.columns:
    df_monthly_crimes = df_street.groupby('Month')[substrings].sum().reset_index()
```

```python
# (Make sure df_monthly_crimes is the DataFrame created above)

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

crime_count_columns = df_monthly_crimes[substrings] # Select only the crime count columns
correlation_matrix = crime_count_columns.corr(method='pearson') # Or 'pearson'

# Seaborn heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap of Crime Types (Spearman)')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
